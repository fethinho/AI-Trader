````python
# KÜTÜPHANELER
!pip install yfinance pandas_ta -q

import yfinance as yf
import pandas as pd
import pandas_ta as ta

# 1) BIST / IBIST hisse listesi (sabit liste)
tickers = [
 "A1CAP.IS", "A1YEN.IS", "ACSEL.IS", "ADEL.IS", "ADESE.IS", "ADGYO.IS", "AEFES.IS", "AFYON.IS",
 "AGESA.IS", "AGHOL.IS", "AGROT.IS", "AHGAZ.IS", "AHSGY.IS", "AKBNK.IS", "AKCNS.IS", "AKENR.IS",
 "AKFIS.IS", "AKFYE.IS", "AKGRT.IS", "AKSA.IS", "AKSEN.IS", "AKSUE.IS", "AKYHO.IS", "ALARK.IS",
 "ALBRK.IS", "ALCAR.IS", "ALCTL.IS", "ALFAS.IS", "ALKA.IS", "ALKIM.IS", "ALKLC.IS",
 "ALTNY.IS", "ALVES.IS", "ANELE.IS", "ANGEN.IS", "ANHYT.IS", "ANSGR.IS", "ARASE.IS", "ARCLK.IS",
 "ARDYZ.IS", "ARENA.IS",  "ARMGD.IS", "ARSAN.IS", "ARTMS.IS", "ARZUM.IS", "ASELS.IS",
 "ASTOR.IS", "ASUZU.IS", "ATAKP.IS", "ATATP.IS", "ATEKS.IS", "ATLAS.IS", "ATSYH.IS", "AVGYO.IS",
 "AVHOL.IS", "AVOD.IS", "AVPGY.IS", "AYCES.IS", "AYDEM.IS", "AYEN.IS", "AYES.IS", "AYGAZ.IS",
 "AZTEK.IS", "BAGFS.IS", "BAHKM.IS", "BAKAB.IS", "BALAT.IS", "BALSU.IS", "BANVT.IS", "BARMA.IS",
 "BASCM.IS", "BASGZ.IS", "BAYRK.IS", "BEGYO.IS",  "BESLR.IS", "BEYAZ.IS", "BFREN.IS",
 "BIENY.IS", "BIGCH.IS", "BIGEN.IS", "BIGTK.IS", "BIMAS.IS", "BINBN.IS", "BINHO.IS", "BIOEN.IS",
 "BIZIM.IS", "BJKAS.IS", "BLCYT.IS", "BLUME.IS", "BMSCH.IS", "BMSTL.IS", "BNTAS.IS", "BOBET.IS",
 "BORLS.IS", "BORSK.IS", "BOSSA.IS", "BRISA.IS", "BRKO.IS", "BRKSN.IS", "BRKVY.IS", "BRLSM.IS",
 "BRMEN.IS", "BRSAN.IS", "BRYAT.IS", "BSOKE.IS", "BTCIM.IS", "BUCIM.IS", "BULGS.IS", "BURCE.IS",
 "BURVA.IS", "BVSAN.IS", "BYDNR.IS", "CANTE.IS", "CASA.IS", "CATES.IS", "CCOLA.IS", "CELHA.IS",
 "CEMAS.IS", "CEMTS.IS", "CEMZY.IS", "CEOEM.IS", "CGCAM.IS", "CIMSA.IS", "CLEBI.IS", "CMBTN.IS",
 "CMENT.IS", "CONSE.IS", "COSMO.IS", "CRDFA.IS", "CRFSA.IS", "CUSAN.IS", "CVKMD.IS", "CWENE.IS",
 "DAGI.IS", "DAPGM.IS", "DARDL.IS", "DCTTR.IS", "DENGE.IS", "DERHL.IS", "DERIM.IS", "DESA.IS",
 "DESPC.IS", "DEVA.IS", "DGATE.IS", "DGNMO.IS", "DIRIT.IS", "DITAS.IS", "DMRGD.IS", "DMSAS.IS",
 "DNISI.IS", "DOAS.IS", "DOFER.IS", "DOFRB.IS", "DOGUB.IS", "DOHOL.IS", "DOKTA.IS", "DSTKF.IS",
 "DUNYH.IS", "DURDO.IS", "DURDO.IS", "DURKN.IS", "DYOBY.IS", "DZGYO.IS", "EBEBK.IS", "ECILC.IS",
 "ECZYT.IS", "EDATA.IS", "EDIP.IS", "EFOR.IS", "EGEEN.IS", "EGEGY.IS", "EGPRO.IS", "EGSER.IS",
 "EKIZ.IS","GRTHO.IS", "GATEG.IS",
 "EKOS.IS", "EKSUN.IS", "ELITE.IS", "EMKEL.IS", "EMNIS.IS", "ENKAI.IS", "ENSRI.IS", "ENTRA.IS",
 "EPLAS.IS", "ERBOS.IS", "ERCB.IS", "EREGL.IS", "ERSU.IS", "ESCAR.IS", "ESCOM.IS", "ESEN.IS",
 "ETILR.IS", "ETYAT.IS", "EUHOL.IS", "EUKYO.IS", "EUPWR.IS", "EUREN.IS", "EUYO.IS", "FADE.IS",
 "FENER.IS", "FLAP.IS", "FMIZP.IS", "FONET.IS", "FORMT.IS", "FORTE.IS", "FRIGO.IS", "FROTO.IS",
 "FZLGY.IS", "GARAN.IS", "GARFA.IS", "GEDIK.IS", "GEDZA.IS", "GENIL.IS", "GENTS.IS", "GEREL.IS",
 "GESAN.IS", "GIPTA.IS", "GLBMD.IS", "GLCVY.IS", "GLRYH.IS", "GLYHO.IS", "GMTAS.IS", "GOKNR.IS",
  "GSRAY.IS", "GUBRF.IS", "GUNDG.IS", "GWIND.IS", "GZNMI.IS", "HALKB.IS", "HATEK.IS", "HATSN.IS",
 "HDFGS.IS", "HEDEF.IS", "HEKTS.IS", "HKTM.IS", "HOROZ.IS", "HRKET.IS", "HTTBT.IS", "HUBVC.IS",
 "HUNER.IS", "HURGZ.IS", "ICBCT.IS", "ICUGS.IS", "IEYHO.IS", "IHAAS.IS", "IHEVA.IS", "IHGZT.IS",
 "IHLAS.IS", "IHLGM.IS", "IHYAY.IS", "IMASM.IS", "INDES.IS", "INFO.IS", "INGRM.IS", "INTEK.IS",
 "INTEM.IS", "INVEO.IS", "INVES.IS", "ISBIR.IS", "ISDMR.IS", "ISFIN.IS", "ISKPL.IS", "ISMEN.IS",
 "ISSEN.IS", "IZENR.IS", "IZFAS.IS", "IZINV.IS", "IZMDC.IS", "JANTS.IS", "KAPLM.IS", "KAREL.IS",
 "KARSN.IS", "KARTN.IS", "KATMR.IS", "KAYSE.IS", "KBORU.IS", "KCAER.IS", "KCHOL.IS", "KENT.IS",
 "KERVN.IS", "KFEIN.IS", "KIMMR.IS", "KLKIM.IS", "KLMSN.IS", "KLNMA.IS", "KLRHO.IS", "KLSER.IS",
 "KLSYN.IS", "KLYPV.IS", "KMPUR.IS", "KNFRT.IS", "KOCMT.IS", "KONKA.IS", "KONTR.IS", "KONYA.IS",
 "KOPOL.IS", "KORDS.IS", "KOTON.IS", "KRDMD.IS", "KRONT.IS", "KRPLS.IS", "KRSTL.IS", "KRTEK.IS",
 "KRVGD.IS", "KSTUR.IS", "KTLEV.IS", "KTSKR.IS", "KUTPO.IS", "KUVVA.IS", "KZBGY.IS", "KZGYO.IS",
 "LIDER.IS", "LIDFA.IS", "LILAK.IS", "LINK.IS", "LKMNH.IS", "LMKDC.IS", "LOGO.IS", "LRSHO.IS",
 "LUKSK.IS", "LYDHO.IS", "LYDYE.IS", "MAALT.IS", "MACKO.IS", "MAGEN.IS", "MAKIM.IS", "MAKTK.IS",
 "MANAS.IS", "MARBL.IS", "MARKA.IS", "MARMR.IS", "MARTI.IS", "MAVI.IS", "MEDTR.IS", "MEGAP.IS",
 "MEGMT.IS", "MEKAG.IS", "MEPET.IS", "MERCN.IS", "MERIT.IS", "MERKO.IS", "METRO.IS", "MGROS.IS",
 "MHRGY.IS", "MIATK.IS", "MMCAS.IS", "MNDRS.IS", "MNDTR.IS", "MOBTL.IS", "MOGAN.IS", "MOPAS.IS",
 "MPARK.IS", "MRSHL.IS", "MTRKS.IS", "MTRYO.IS", "MZHLD.IS", "NATEN.IS", "NETAS.IS", "NIBAS.IS",
 "NTGAZ.IS", "NTHOL.IS", "NUHCM.IS", "OBAMS.IS", "OBASE.IS", "ODAS.IS", "ODINE.IS", "OFSYM.IS",
 "ONCSM.IS", "ONRYT.IS", "ORCAY.IS", "ORGE.IS", "ORMA.IS", "OSMEN.IS", "OSTIM.IS", "OTKAR.IS",
 "OTTO.IS", "OYAKC.IS", "OYAYO.IS", "OYLUM.IS", "OYYAT.IS", "OZATD.IS", "OZRDN.IS", "OZYSR.IS",
 "PAGYO.IS", "PAMEL.IS", "PAPIL.IS", "PARSN.IS", "PASEU.IS", "PATEK.IS", "PCILT.IS", "PEKGY.IS",
 "PENGD.IS", "PENTA.IS", "PETKM.IS", "PETUN.IS", "PGSUS.IS", "PINSU.IS", "PKART.IS", "PKENT.IS",
 "PLTUR.IS", "PNLSN.IS", "PNSUT.IS", "POLHO.IS", "POLTK.IS", "PRDGS.IS", "PRKAB.IS", "PRKME.IS",
 "PRZMA.IS", "PSDTC.IS", "QNBFK.IS", "QNBTR.IS", "QUAGR.IS", "RALYH.IS", "RAYSG.IS", "REEDR.IS",
 "RGYAS.IS", "RNPOL.IS", "RODRG.IS", "RTALB.IS", "RUBNS.IS", "RUZYE.IS", "RYSAS.IS", "SAFKR.IS",
 "SAHOL.IS", "SAMAT.IS", "SANEL.IS", "SANFM.IS", "SANKO.IS", "SARKY.IS", "SASA.IS", "SAYAS.IS",
 "SDTTR.IS", "SEGYO.IS", "SEKFK.IS", "SEKUR.IS", "SELVA.IS", "SERNT.IS", "SEYKM.IS", "SILVR.IS",
 "SISE.IS", "SKBNK.IS", "SKTAS.IS", "SKYMD.IS", "SKYLP.IS", "SMART.IS", "SMRTG.IS", "SMRVA.IS",
 "SNICA.IS",  "SNPAM.IS", "SODSN.IS", "SOKE.IS", "SOKM.IS", "SONME.IS", "SRVGY.IS",
 "SUMAS.IS", "SUNTK.IS", "SURGY.IS", "SUWEN.IS", "TABGD.IS", "TARKM.IS", "TATEN.IS", "TATGD.IS",
 "TAVHL.IS", "TBORG.IS", "TCELL.IS", "TCKRC.IS", "TDGYO.IS", "TRMET.IS", "TEZOL.IS", "TGSAS.IS",
 "THYAO.IS", "TKFEN.IS", "TKNSA.IS", "TLMAN.IS", "TMPOL.IS", "TMSN.IS", "TNZTP.IS", "TOASO.IS",
 "TRALT.IS", "TRCAS.IS", "TRENJ.IS", "TRHOL.IS", "TRILC.IS", "TRMET.IS", "TSKB.IS", "TSPOR.IS",
 "TTKOM.IS", "TTRAK.IS", "TUCLK.IS", "TUKAS.IS", "TUPRS.IS", "TUREX.IS", "TURGG.IS", "TURSG.IS",
 "UFUK.IS", "ULAS.IS", "ULKER.IS", "ULUFA.IS", "ULUSE.IS", "ULUUN.IS", "UNLU.IS", "USAK.IS",
 "VAKBN.IS", "VAKFA.IS", "VAKFN.IS", "VAKKO.IS", "VANGD.IS", "VBTYZ.IS", "VERUS.IS", "VESBE.IS",
 "VESTL.IS", "VKFYO.IS", "VKING.IS", "VRGYO.IS", "VSNMD.IS", "YAPRK.IS", "YATAS.IS", "YAYLA.IS",
 "YBTAS.IS", "YEOTK.IS", "YESIL.IS", "YGGYO.IS", "YIGIT.IS", "YKBNK.IS", "YKSLN.IS", "YONGA.IS",
 "YUNSA.IS", "YYAPI.IS", "YYLGD.IS", "ZEDUR.IS", "ZOREN.IS", "ZRGYO.IS"
]

uygun_hisseler = []

# *** Tarama aralığını belirleyin ***
# Geçerli değerler: "1m", "2m", "5m", "15m", "30m", "60m", "90m", "1h", "5h", "1d", "5d", "1wk", "1mo", "3mo"
# Not: yfinance bazı kısa aralıklar için sınırlı geçmiş veri sağlar (örn: 30m için max 60 gün)
scan_interval = "1d"  # BURAYI İSTEDİĞİNİZ TARAMA ARALIĞI İLE DEĞİŞTİRİN

# Tarama aralığına göre veri indirme periyodunu ayarla
if scan_interval in ["1m", "2m", "5m", "15m", "30m", "60m", "90m", "1h", "5h"]:
    download_period = "60d"  # Kısa aralıklar için daha kısa geçmiş
elif scan_interval in ["1d"]:
    download_period = "1y"   # Günlük için 1 yıl yeterli
elif scan_interval in ["1wk", "1mo", "3mo"]:
    download_period = "5y"   # Haftalık/aylık için daha uzun geçmiş
else:
    download_period = "1y" # Varsayılan olarak 1 yıl

print(f"Toplam taranacak hisse sayısı: {len(tickers)}")

# Yeni adım: Hata veren hisse sembollerini temizleme
print("\nGeçersiz veya veri bulunamayan hisse senetleri taranıyor ve ana listeden çıkarılıyor...")
filtered_tickers = []
for ticker in tickers:
    try:
        # Sadece bir günlük veri çekerek hissenin geçerliliğini hızlıca kontrol et
        # Auto_adjust=True ile bölünme ve temettü ayarlamalarını otomatik yapar
        test_df = yf.download(ticker, period="1d", interval="1d", progress=False, auto_adjust=True)
        if not test_df.empty:
            filtered_tickers.append(ticker)
        else:
            print(f"❌ '{ticker}' sembolü için veri bulunamadı veya geçersiz. Listeden çıkarılıyor.")
    except Exception as e:
        print(f"❌ '{ticker}' sembolü indirilirken hata oluştu ({e}). Listeden çıkarılıyor.")

if len(filtered_tickers) < len(tickers):
    print(f"\n{len(tickers) - len(filtered_tickers)} adet geçersiz hisse senedi elendi.")

tickers = filtered_tickers

if not tickers:
    print("Hisse senedi listesinde hiç geçerli sembol kalmadı. Analiz yapılamıyor.")
    # Eğer geçerli hisse kalmazsa, sonraki döngü çalışmayacaktır.
else:
    print(f"Kalan geçerli hisse senedi sayısı: {len(tickers)}")

print(f"\nHisseler '{scan_interval}' aralığı ve '{download_period}' periyodu ile taranıyor, lütfen bekleyin...\n")

for ticker in tickers:
    try:
        # EMA(50) ve 10 günlük hacim ortalaması için en az 100 günlük veri çekiyoruz
        # Veri periyodu ve aralığı dinamik olarak ayarlandı
        df = yf.download(ticker, period=download_period, interval=scan_interval, progress=False, auto_adjust=True)

        if df.empty or len(df) < 50:
            continue

        if isinstance(df.columns, pd.MultiIndex):
            df.columns = df.columns.droplevel(1)

        df["EMA_20"] = ta.ema(df["Close"], length=20)
        df["EMA_50"] = ta.ema(df["Close"], length=50)
        df["RSI_14"] = ta.rsi(df["Close"], length=14)
        df["Vol_Ort_10"] = ta.sma(df["Volume"], length=10)

        if len(df) < 2:
            continue

        son_gun = df.iloc[-1]
        onceki_gun = df.iloc[-2]

        fiyat = son_gun["Close"]
        ema_20_bugun = son_gun["EMA_20"]
        ema_50_bugun = son_gun["EMA_50"]
        ema_20_dun = onceki_gun["EMA_20"]
        ema_50_dun = onceki_gun["EMA_50"]
        rsi_14 = son_gun["RSI_14"]
        hacim = son_gun["Volume"]
        hacim_ort_10 = son_gun["Vol_Ort_10"]

        if (
            pd.isna(ema_50_bugun) or pd.isna(ema_20_bugun) or
            pd.isna(ema_50_dun)   or pd.isna(ema_20_dun)   or
            pd.isna(rsi_14)       or pd.isna(hacim_ort_10)
        ):
            continue

        # STRATEJİ KOŞULLARI
        kosul_1 = fiyat >= ema_20_bugun           # Fiyat >= EMA(20) (bugün)
        kosul_2 = 40 < rsi_14 < 89               # 40 < RSI(14) < 89
        kosul_3 = hacim > hacim_ort_10           # Hacim > 10 günlük ortalama
        # EMA20'nin EMA50'yi BUGÜN yukarı kesmesi:
        # Dün: EMA20 <= EMA50, Bugün: EMA20 > EMA50
        kosul_4 = (ema_20_dun <= ema_50_dun) and (ema_20_bugun > ema_50_bugun)

        if kosul_1 and kosul_2 and kosul_3 and kosul_4:
            uygun_hisseler.append({
                "Hisse": ticker,
                "Fiyat": round(fiyat, 2),
                "RSI(14)": round(rsi_14, 2),
                "EMA(20)_Bugun": round(ema_20_bugun, 2),
                "EMA(50)_Bugun": round(ema_50_bugun, 2),
                "EMA(20)_Dun": round(ema_20_dun, 2),
                "EMA(50)_Dun": round(ema_50_dun, 2),
                "Hacim": int(hacim),
                "Ort. Hacim(10G)": int(hacim_ort_10),
            })

    except Exception as e:
        print(f"Hata oluştu ({ticker}): {e}")

if uygun_hisseler:
    print(f"Belirttiğiniz koşulları sağlayan (EMA20 bugün EMA50’yi YUKARI kesen) BIST hisseleri ('{scan_interval}' aralığı için):\n")
    sonuc_df = pd.DataFrame(uygun_hisseler)
    display(sonuc_df)

    sonuc_df.to_csv("/content/sonuc_bist_taramasi.csv", index=False, encoding="utf-8-sig")
    print("\nSonuçlar 'sonuc_bist_taramasi.csv' dosyasına kaydedildi.")
else:
    print(f"Mevcut durumda koşulları sağlayan (EMA20 bugün EMA50’yi yukarı kesen) hisse ('{scan_interval}' aralığı için) bulunamadı.")
````

```yaml
    # Her hafta içi (Pazartesi-Cuma) UTC 05:30'da çalışır (Türkiye saatiyle 08:30)
    # UTC+3 = 08:30 -> UTC = 05:30
    - cron: '30 5 * * 1-5'
  workflow_dispatch: # Manuel çalıştırmaya izin verir

jobs:
  scan-and-send:
    runs-on: ubuntu-latest

    steps:
    - name: Checkout repository
      uses: actions/checkout@v4

    - name: Set up Python
      uses: actions/setup-python@v5
      with:
        python-version: '3.x'

    - name: Install dependencies
      run: |
        python -m pip install --upgrade pip
        pip install -r requirements.txt

    - name: Run Stock Scan and Send to Telegram
      run: python main.py
      env:
        TELEGRAM_BOT_TOKEN: ${{ secrets.TELEGRAM_BOT_TOKEN }}
        TELEGRAM_CHAT_ID: ${{ secrets.TELEGRAM_CHAT_ID }}
```
